```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Set up the screen
screen_width = 800
screen_height = 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Catch the Falling Objects")

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)

# Player and object attributes
player_width = 80
player_height = 20
player_x = (screen_width - player_width) / 2
player_y = screen_height - player_height - 10
player_speed = 10

object_width = 30
object_height = 30
object_speed = 5
objects = []

# Font settings
font = pygame.font.Font(None, 36)

# Game loop
running = True
score = 0
clock = pygame.time.Clock()

while running:
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Move the player
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and player_x > 0:
        player_x -= player_speed
    if keys[pygame.K_RIGHT] and player_x < screen_width - player_width:
        player_x += player_speed

    # Spawn new objects
    if len(objects) < 5:
        new_object_x = random.randint(0, screen_width - object_width)
        new_object = [new_object_x, 0, object_width, object_height]
        objects.append(new_object)

    # Move objects
    for obj in objects:
        obj[1] += object_speed

    # Check for collisions
    for obj in objects:
        if obj[1] + obj[3] >= screen_height:
            objects.remove(obj)
        elif player_x <= obj[0] <= player_x + player_width or player_x <= obj[0] + obj[2] <= player_x + player_width:
            if player_y <= obj[1] <= player_y + player_height or player_y <= obj[1] + obj[3] <= player_y + player_height:
                objects.remove(obj)
                score += 1

    # Clear the screen
    screen.fill(WHITE)

    # Draw the player
    pygame.draw.rect(screen, BLACK, (player_x, player_y, player_width, player_height))

    # Draw the objects
    for obj in objects:
        pygame.draw.rect(screen, RED, obj)

    # Draw the score
    score_text = font.render("Score: " + str(score), True, BLACK)
    screen.blit(score_text, (10, 10))

    # Update the display
    pygame.display.flip()

    # Set the frame rate
    clock.tick(60)

# Quit Pygame
pygame.quit()
```
